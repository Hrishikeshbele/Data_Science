```python
# importing necessary libraries
from csv import reader
from random import seed
from random import randrange
from math import sqrt
from math import exp
from math import pi
```

```python
#importing CMC dataset from csv file
def load_csv(filename):
    lines=reader(open(filename, 'r'))
    dataset=list(lines)
    for i in range(len(dataset)):
        dataset[i]=[float(x) for x in dataset[i]]
    return dataset
```

```python
# Split the dataset by class , returns a dictionary
#no of instances of each class: class1-629 ,class2-333, class3- 511
def separate_by_class(dataset):
    separated = dict()
    for i in range(len(dataset)):
        vector = dataset[i]
        class_value = vector[-1]
        if (class_value not in separated):
            separated[class_value] = list()
        separated[class_value].append(vector)
    return separated
```

```python
#We need two statistics from a given set of data for continous variable

# function to Calculate the mean of a list of numbers
def mean(numbers):
    return float(sum(numbers))/float(len(numbers))

# function to Calculate the standard deviation of a list of numbers
def stdev(numbers):
    avg = mean(numbers)
    variance = sum([(x-avg)**2 for x in numbers]) / float(len(numbers)-1)
    return sqrt(variance)

# function to Calculate the mean, stdev and count for each column in a dataset
def summarize_dataset(dataset):
    summaries = [(mean(column), stdev(column), len(column)) for column in zip(*dataset)]
    del(summaries[-1])
    return summaries

# function to Split dataset by class then calculate statistics for each column
def summarize_by_class(dataset):
    separated = separate_by_class(dataset)
    summaries = dict()
    for class_value, rows in separated.items():
        summaries[class_value] = summarize_dataset(rows)
    return summaries
```

```python
# k-fold cross-validation step

# Split a dataset into k folds
def cross_validation_split(dataset, k_folds):
    dataset_split = list()
    dataset_copy = list(dataset)
    fold_size = int(len(dataset) / k_folds)
    for _ in range(k_folds):
        fold = list()
        while len(fold) < fold_size:
            index = randrange(len(dataset_copy))
            fold.append(dataset_copy.pop(index))
        dataset_split.append(fold)
    return dataset_split

# Evaluate an algorithm using a cross validation split
def evaluate_algorithm(dataset, algorithm, k_folds, *args):
    folds = cross_validation_split(dataset, k_folds)
    scores = list()
    for fold in folds:
        train_set = list(folds)
        train_set.remove(fold)
        train_set = sum(train_set, [])
        test_set = list()
        for row in fold:
            row_copy = list(row)
            test_set.append(row_copy)
            row_copy[-1] = None
        predicted = algorithm(train_set, test_set, *args)
        actual = [row[-1] for row in fold]
        conf_matrix=create_conf_matrix(actual, predicted, 3)
        print('new fold result')
        print('confusion matrix:',conf_matrix)
        pred_matric(conf_matrix)
        accuracy = accuracy_metric(actual, predicted)
        scores.append(accuracy)
    return scores
```

```python
#funtions to calculate vaious evaluation matrices

#funtion to calculate confusion matrix
def create_conf_matrix(expected, predicted, n_classes):
    m = [[0] * n_classes for i in range(n_classes)]
    for pred, exp in zip(predicted, expected):
        #print(pred,exp)
        m[int(pred)-1][int(exp)-1] += 1
    return m

#funtion to calculate precision,recall and f1 score
def pred_matric(conf_matrix):
    t = sum(sum(l) for l in conf_matrix)
    for i in range(len(conf_matrix)):
        precision=conf_matrix[i][i] /sum(conf_matrix[i][j]for j in range(len(conf_matrix)))
        recall= conf_matrix[i][i]/sum(conf_matrix[j][i] for j in range(len(conf_matrix)))
        f1=2*precision*recall/(precision+recall)
        print('class: ',i+1,'precision: %s' % round(precision,2),'recall: %s' % round(recall,2),'f1: %s' % round(f1,2))
        
# function to Calculate accuracy percentage
def accuracy_metric(actual, predicted):
    correct = 0
    for i in range(len(actual)):
        if actual[i] == predicted[i]:
            correct += 1
    return round(correct / float(len(actual)),2 )

```

```python
#calculates the probabilities of a new instance belonging to each class and selecting the class with the largest probability value.
def predict(summaries, row):
    probabilities = calculate_posterior_probabilities(summaries, row)
    best_label, best_prob = None, -1
    for class_value, probability in probabilities.items():
        if best_label is None or probability > best_prob:
            best_prob = probability
            best_label = class_value
    return best_label
 
```

```python
# Naive Bayes Algorithm
def naive_bayes(train, test):
    summarize = summarize_by_class(train)
    predictions = list()
    for row in test:
        output = predict(summarize, row)
        predictions.append(output)
    return(predictions)
 
# Test Naive Bayes on MCM Dataset
seed(10)
filename = (r"D:\DS_challenges\CS4003 DA Projects\CMC.csv")
dataset = load_csv(filename)
# evaluate algorithm using k fold cross validation
n_folds = 5
scores = evaluate_algorithm(dataset[:], naive_bayes, n_folds)
print('accuracy Scores: %s' % scores)
print('Mean Accuracy: %.3f%%' % (sum(scores)/float(len(scores))))
```
